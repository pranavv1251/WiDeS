```python
data = np.load('/content/gdrive/MyDrive/Pranav WiDes Data/Data Processing/Output Files/Copy of path_bld01_tx001_ouput.npz')['data']
data, data.shape
data[0,0,0]
```

```python
values_to_extract = np.zeros((29929, 45, 2), dtype=np.complex128)
values_to_extract[:, :, 0] = data[:, :, 2, 0]  # Extract values at position [2, 0]
values_to_extract[:, :, 1] = data[:, :, 2, 1]

```

```python
values_to_extract[np.isnan(values_to_extract)] = 0 + 0.j

values_to_extract.shape
```

```python
theta = np.random.uniform(0, np.pi, (29929,45))
theta.shape
```

```python
Pi = values_to_extract[:, :, 0]
delay = values_to_extract[:, :, 1]
power = 10 ** ((Pi - 30) / 20)

# print(power)
power.shape
```

```python
comp = np.zeros((29929,45 ), dtype=np.complex128)

comp = np.exp(1j * theta)

power = power * comp
power.shape
```

```python
M = 200
bw = 400
freq = np.linspace(-bw/2, bw/2 - bw/M, M)
freq_reshaped = freq.reshape(200, 1, 1)
freq.shape, freq_reshaped.shape
```

```python
h = np.exp(1j * 2 * np.pi * freq_reshaped * delay)
a = power * h
a.shape
```
